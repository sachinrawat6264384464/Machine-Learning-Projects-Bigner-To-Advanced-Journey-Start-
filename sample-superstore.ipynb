```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data file from folder by the help of pandas
df=pd.read_csv("Sample Superstore.csv", encoding="latin1")
df.head()
```

```python
df = df.drop_duplicates(subset=["Order ID"])
```

```python
x=df.drop(['Row ID',  'Ship Date', 'Ship Mode',
        'Customer Name',  'Country', 
       'Postal Code', 'Region', 'Product ID', 
       'Product Name'],axis=1)
```

```python
x["year"]=x["Order Date"].dt.year
```

```python
x.shape
```

```python
max_date=x["Order Date"].max()
print("maxing",max_date)
```

```python
rfm=x.groupby(['Customer ID','City','Segment']).agg({
    'Order Date': lambda x: (max_date - x.max()).days,  # Recency
    'Order ID': 'count',                                # Frequency
    'Sales': 'sum',                                     # Monetary (Sales)
    'Profit': 'sum'                                     # Monetary (Profit)
}).reset_index()
```

```python
rfm.columns=['Customer ID','City','Segment', 'Recency', 'Frequency', 'Monetary_Sales', 'Monetary_Profit']
rfm.head()
```

```python
rfm['Risky_Customer'] = np.where(
    (rfm['Monetary_Sales'] > rfm['Monetary_Sales'].median()) & 
    (rfm['Monetary_Profit'] <= 0),
    "Yes",
    "No"
)

rfm[rfm['Risky_Customer'] == "Yes"].head(20)
```

```python
rfm['Monetary_Sales'] = np.log1p(rfm['Monetary_Sales'])
rfm.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary_Sales']])

rfm_scaled_df = pd.DataFrame(rfm_scaled, columns=['Recency', 'Frequency', 'Monetary_Sales'])
rfm_scaled_df.head()
```

```python
churn_threshold = 90

rfm['Churn_Flag'] = rfm['Recency'].apply(lambda x: 1 if x > churn_threshold else 0)

rfm[['Customer ID','Recency','Churn_Flag']].head()
```

```python
churn_rate = (rfm['Churn_Flag'].sum() / len(rfm)) * 100
print("Churn Rate:", churn_rate)
```

```python
threshold = rfm['Recency'].quantile(0.75)

rfm['Churn_Flag'] = (rfm['Recency'] > threshold).astype(int)

threshold
```

```python
rfm['Churn_Flag'].value_counts(normalize=True) * 100
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmean.fit(rfm_scaled)
    wcss.append(kmean.inertia_) 
  
```

```python
sns.lineplot(x=range(1,11),y=wcss,marker="o")
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmen=KMeans(n_clusters=5,random_state=42)
rfm["cluster"]=kmen.fit_predict(rfm_scaled)
```

```python
# just find the score in this show that how cluster perform
from sklearn.metrics import silhouette_score
score=silhouette_score(rfm_scaled, rfm["cluster"])
print("Silhouette Score:", score)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# PCA reduce to 2 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(rfm_scaled_df)

# Plot PCA graph
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0], pca_data[:,1], c=rfm['cluster'], cmap='viridis', s=50)

plt.title("PCA Visualization of Customer Clusters")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label="Cluster")
plt.show()
```

```python
centers = kmen.cluster_centers_
centers_original = scaler.inverse_transform(centers)
```

```python

# we make a graph scaterplot to show the clusters
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=rfm['Monetary_Sales'],
    y=rfm['Frequency'],
    hue=rfm['cluster'],
    palette='Set2',
    s=100
)

# Centroids plot (original scale)
plt.scatter(
    centers_original[:,2],   # Monetary centroid
    centers_original[:,1],   # Frequency centroid
    c="black",
    s=250,
    marker="X",
    label="Centroids"
)

plt.title("Customer Segmentation using K-Means (RFM)")
plt.xlabel("Monetary (Total Spend)")
plt.ylabel("Frequency (Total Orders)")
plt.legend() 
plt.show()
```

```python
cluster_summary = rfm.groupby("cluster").agg({
    "Recency": "mean",
    "Frequency": "mean",
    "Monetary_Sales": "mean",
    "Monetary_Profit": "mean",
    "Customer ID": "count"
}).rename(columns={"Customer ID":"Customer_Count"}).reset_index()

cluster_summary
```

```python
cluster_summary["R_rank"] = cluster_summary["Recency"].rank(ascending=True)
cluster_summary["F_rank"] = cluster_summary["Frequency"].rank(ascending=False)
cluster_summary["M_rank"] = cluster_summary["Monetary_Sales"].rank(ascending=False)

cluster_summary["Score"] = cluster_summary["R_rank"] + cluster_summary["F_rank"] + cluster_summary["M_rank"]

cluster_summary = cluster_summary.sort_values("Score")
cluster_summary
```

```python
segment_map = {
    0: "Lost Customers",
    1: "Low Value",
    2: "Champions",
    3: "At Risk",
    4: "Loyal Customers"
}

rfm["Segmentn"] = rfm["cluster"].map(segment_map)
rfm.head(40)
```

```python
segment_sales = rfm.groupby("Segment")["Monetary_Sales"].sum().reset_index()

segment_sales["Revenue_%"] = (segment_sales["Monetary_Sales"] / segment_sales["Monetary_Sales"].sum()) * 100

segment_sales.sort_values(by="Revenue_%", ascending=False)
```

```python
pareto = rfm.groupby("Segment").agg({
    "Customer ID": "count",
    "Monetary_Sales": "sum"
}).reset_index()

pareto["Customer_%"] = (pareto["Customer ID"] / pareto["Customer ID"].sum()) * 100
pareto["Revenue_%"] = (pareto["Monetary_Sales"] / pareto["Monetary_Sales"].sum()) * 100

pareto = pareto.sort_values(by="Revenue_%", ascending=False)

pareto["Cumulative_Revenue_%"] = pareto["Revenue_%"].cumsum()
pareto["Cumulative_Customers_%"] = pareto["Customer_%"].cumsum()

pareto
```

```python
# Segment-wise total sales
segment_sales = rfm.groupby("Segmentn")["Monetary_Sales"].sum().reset_index()

# Total sales
total_sales = segment_sales["Monetary_Sales"].sum()

# Revenue Contribution %
segment_sales["Revenue_Contribution_%"] = (segment_sales["Monetary_Sales"] / total_sales) * 100

# Sort descending
segment_sales = segment_sales.sort_values(by="Revenue_Contribution_%", ascending=False)

segment_sales
```

```python
for index, row in segment_sales.iterrows():
    print(f"{row['Segmentn']} = {row['Revenue_Contribution_%']:.2f}% revenue")
```
