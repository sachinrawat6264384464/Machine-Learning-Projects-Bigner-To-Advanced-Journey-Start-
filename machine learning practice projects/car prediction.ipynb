```python
# import some library for preprocessing and visualizing the data  and saome sklearn libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# read the data

df=pd.read_csv("CarPrice_Assignment.csv")
```

```python
x=df.drop(['car_ID', 'CarName', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight',
        'enginesize', 'boreratio', 
       'compressionratio','price','fuelsystem'],axis=1)
```

```python
x.head()
```

```python
df["fuelsystem"].unique()
```

```python
# in this we convert object columns into intger column
x["fueltype"]=x["fueltype"].map({"gas":1,"diesel":2})
x["aspiration"]=x["aspiration"].map({"std":1,"turbo":2})
x["doornumber"]=x["doornumber"].map({"two":2,"four":4})
x["carbody"]=x["carbody"].map({"convertible":1,"hatchback":2,"sedan":3,"wagon":4,"hardtop":5})
x["drivewheel"]=x["drivewheel"].map({"rwd":1,"fwd":2,"4wd":3})
x["enginelocation"]=x["enginelocation"].map({"front":1,"rear":2})
x["cylindernumber"]=x["cylindernumber"].map({'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8})
x["enginetype"]=x["enginetype"].map({'dohc':1 ,'ohcv':2, 'ohc':3, 'l':4, 'rotor':5, 'ohcf':6, 'dohcv':7})
```

```python
x.head()
```

```python
x.isnull().sum()
```

```python
x.shape
```

```python
# y is for train and test the model this very usefull for model 
y=df["price"]
```

```python
#in this train the data 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
# use a random forest regressor model for regression problem
svc_model=RandomForestRegressor(max_depth= 20, max_features= "sqrt", min_samples_leaf= 1, min_samples_split= 2, n_estimators= 400 ,random_state=42)
pred=svc_model.fit(x_train,y_train)
y_pred=pred.predict(x_test)
```

```python
# checkperformance and data train and test 
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R2 Score:", r2)
```

```python

import numpy as np
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

```python
# this show price value
print("Min:", y.min())
print("Max:", y.max())
print("Mean:", y.mean())
```

```python
# write here i use tuning hyperparameter tuning for selecting best paras_
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [200, 400, 600],
    'max_depth': [20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

# Grid Search
grid_search = GridSearchCV(
    estimator=svc_model,
    param_grid=param_grid,
    cv=3,                # 3-fold cross-validation
    n_jobs=-1,
    verbose=2,
    scoring='neg_mean_squared_error'
)

grid_search.fit(x_train, y_train)

print("Best Parameters:", grid_search.best_params_)
```
