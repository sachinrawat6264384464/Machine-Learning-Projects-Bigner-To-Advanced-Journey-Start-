```python
#practise fradabosting andgriendient boosting from sklearn.ensemble import GradientBoostingClassifierr
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('tested.csv')
df.head()
```

```python
df.drop('PassengerId', axis=1, inplace=True, errors='ignore')
```

```python
df.drop(['Cabin', 'Ticket'], axis=1, inplace=True, errors='ignore')
```

```python
print(df.columns)
```

```python
df['sex1']=df['Sex'].map({'male':1,'female':0})
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
x=df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'sex1']]
y=df['Survived']
```

```python
x.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier( criterion='gini',)
model.fit(X_train,y_train)
train=model.predict(X_train)
test=model.predict(X_test)
print("train accuracy",accuracy_score(y_train,train))
print("test accuracy",accuracy_score(y_test,test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
model1=GradientBoostingClassifier(n_estimators=200,
    learning_rate=0.05,
    max_depth=2,
    random_state=42)
model1.fit(X_train,y_train)
train1=model1.predict(X_train)
test1=model1.predict(X_test)
print("train accuracy",accuracy_score(y_train,train1))
print("test accuracy",accuracy_score(y_test,test1))
```

```python
from sklearn.ensemble import AdaBoostClassifier
model2=AdaBoostClassifier()
model2.fit(X_train,y_train)
train2=model2.predict(X_train)
test2=model2.predict(X_test)
print("train accuracy",accuracy_score(y_train,train2))          
print("test accuracy",accuracy_score(y_test,test2))
```
